# **➲ Aula 06 - Raízes de equações**

## ➥ Uma equação e uma variável
---

Vamos tratar inicialmente dos métodos mais simples para determinar as raízes de uma única equação contendo uma única variável independente. Os métodos mais óbivios podem ser denominados de "Força Bruta". Eles procuram localizar as raízes em sequência a partir da mudança de sinal da função. Vejamos um exemplo.

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

# malha de pontos (valores da variável independente)
t = np.arange(0, 10+1, 0.01)

# nossa função seno
n = 5
y = np.sin(n*t)
```

```python
%matplotlib widget
sns.set()
sns.lineplot(x=t, y=y).set(title=f"y = sen({n}x)")
plt.show()
```

```python
# localiza todos os índices (posições no vetor) que contém a mudança de sinal da função
idx = np.where(np.diff(np.sign(y)))[0]
```

```python
i1 = idx
i2 = idx+1
x1 = t[i1]
x2 = t[i2]
y1 = y[i1]
y2 = y[i2]
# ponto que corta o eixo x (raíz aproximada)
x = x1 - y1*(x2 - x1)/(y2 - y1)
print(x)
```

> Note que encontramos todas as raízes no domínio desejado sem escrever nenhum loop (comando de repetição explícito)!
